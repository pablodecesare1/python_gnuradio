# Generando un float32

```python
samp_rate=32e3
N=1000

ts=(1/samp_rate)
x=np.linspace(0,np.pi*2,N)  # de 0 a 2pi 1000 puntos
t=np.linspace(0,N*ts,N)
y=np.sin(x,dtype="float32")    # algo que hasta ahora no de dabamos importancia cuando creabamos
                               # numpy arrays, el campo de dtype=float32 para ser compatibles con GRC.

plt.plot(t*1000,y)             # ploteo
plt.xlabel("ms")               # multiplicando por 1000 eje de tiempo queda en ms
```

```python
f=open("onda.dat",'wb')
np.ndarray.tofile(y,f,sep="",format="%f")
f.close()
```

La escala de tiempo del eje horizontal indica que el periodo es 31,25ms que corresponde al 1000puntos/32kHz

- Analicemos el espectro

Como la variable "samp_rate" de GNURadio esta configurada en 32kSps (el defaul cuando abrimos el GRC) y tenemos 1 periodo de se単al en 1000 puntos

ts=1/samp_rate= 31,25us

T[peridodo de la se単al]= ts.N= 31.25us . 1000 = 31,25ms

f=1/T=1/31.25ms = 32Hz

## Volvemos a leer la forma de onda generada

```python
a=sc.fromfile(open("onda.dat"),dtype=sc.float32)

### No hay informacion de tiempo por lo tanto debemos generar el vector de tiempo a partir del samp_rate y del largo del vector


t=np.linspace(0,ts*len(a),len(a))
plt.figure() 
plt.plot(t*1000,y)             # ploteo
plt.xlabel("ms")               # multiplicando por 1000 eje de tiempo queda en ms
```

# Generando Complex

```python
samp_rate=32e3
N=1024

ts=(1/samp_rate)

x=np.linspace(0,np.pi*2,N)  # de 0 a 2pi 1000 puntos
t=np.linspace(0,N*ts,N)

c=np.exp(1j*x,dtype="complex64")
plt.plot(t,c.real,t,c.imag)


f=open("onda_complex.dat","wb")
np.ndarray.tofile(c,f,sep="",format="%f")
f.close()

```

## Volvemos a leer la forma de onda generada

```python
samp_rate=32e3
ts=(1/samp_rate)


a=sc.fromfile(open("onda_complex.dat"),dtype=sc.complex64)
t=np.linspace(0,len(a)*ts,len(a))

plt.figure()
plt.plot(t,a.real,t,a.imag)
```
