# Generando un float32

```python
samp_rate=32e3
N=1000

ts=(1/samp_rate)
x=np.linspace(0,np.pi*2,N)  # de 0 a 2pi 1000 puntos
t=np.linspace(0,N*ts,N)
y=np.sin(x,dtype="float32")    # algo que hasta ahora no de dabamos importancia cuando creabamos
                               # numpy arrays, el campo de dtype=float32 para ser compatibles con GRC.

plt.plot(t*1000,y)             # ploteo
plt.xlabel("ms")               # multiplicando por 1000 eje de tiempo queda en ms
```

```python
f=open("onda.dat",'wb')
np.ndarray.tofile(y,f,sep="",format="%f")
f.close()
```

# Generando Complex

```python
samp_rate=32e3
N=1024

ts=(1/samp_rate)

x=np.linspace(0,np.pi*2,N)  # de 0 a 2pi 1000 puntos
t=np.linspace(0,N*ts,N)

c=np.exp(1j*x,dtype="complex64")
plt.plot(t,c.real,t,c.imag)


f=open("onda_complex.dat","wb")
np.ndarray.tofile(c,f,sep="",format="%f")
f.close()

```
